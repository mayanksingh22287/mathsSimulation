```python
# ques2
import numpy as np
import matplotlib.pyplot as plt

# Set the number of experiments
num_experiments = 50000

# Generate four random values for alpha
alpha_values = np.random.uniform(0, 100, 4)

for alpha in alpha_values:
    # Choose the smallest n that provides a reasonable approximation
    n = int(alpha)
    
    # Generate Binomial random variable
    p = alpha - int(alpha)  # Calculate the probability parameter
    binomial_rv = np.random.binomial(n, p, num_experiments)
    
    # Generate Poisson random variable
    poisson_rv = np.random.poisson(alpha, num_experiments)
    
    # Plot the histograms
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.hist(binomial_rv, bins=max(binomial_rv)-min(binomial_rv)+1, density=True, alpha=0.5, label='Binomial')
    plt.title('Binomial (n={}, p={:.2f})'.format(n, p))
    plt.xlabel('Value')
    plt.ylabel('Probability')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.hist(poisson_rv, bins=max(poisson_rv)-min(poisson_rv)+1, density=True, alpha=0.5, label='Poisson')
    plt.title('Poisson (lambda={:.2f})'.format(alpha))
    plt.xlabel('Value')
    plt.ylabel('Probability')
    plt.legend()

    plt.tight_layout()
    plt.show()

```

```python
# ques3
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, poisson

# Set the random seed for reproducibility
np.random.seed(42)

# Set the lambda values
lambda_values = np.random.uniform(1, 20, 2)

# Set the sequence of sample sizes
sample_sizes = [5,10, 100,1000]

for lam in lambda_values:
    plt.figure(figsize=(12, 4))
    
    # Simulate and plot the histogram for each sample size
    for i, sample_size in enumerate(sample_sizes):
        # Generate the Poisson random variables
        poisson_rv = poisson.rvs(lam, size=(sample_size, sample_size))
        
        # Calculate the sums along the rows
        sums = np.sum(poisson_rv, axis=1)
        
        # Plot the histograms
        plt.subplot(2, len(sample_sizes), i + 1)
        plt.hist(sums, bins='auto', density=True, alpha=0.5)
        plt.title('Sample Size: {}'.format(sample_size))
        plt.xlabel('Sum of Poisson RVs')
        plt.ylabel('Probability')
        
        # Plot the corresponding normal distribution for comparison
        mean = sample_size * lam
        std = np.sqrt(sample_size * lam)
        x = np.linspace(mean - 3 * std, mean + 3 * std, 100)
        plt.plot(x, norm.pdf(x, mean, std), color='r', linewidth=2)
        
    plt.tight_layout()
    plt.show()

```


The central limit theorem says that the sampling distribution of the mean will always be normally distributed, as long as the sample size is large enough. Regardless of whether the population has a normal, Poisson, binomial, or any other distribution, the sampling distribution of the mean will be normal.
